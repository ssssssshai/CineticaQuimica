```python
import pygame
import numpy as np
import sys
import matplotlib.pyplot as plt
from scipy import stats
import random  # Import random module

# Initialize Pygame
pygame.init()

# Simulation parameters
num_particles = 100  # Total number of particles
box_width = 800
box_height = 600
radius_A = 10  # Radius for particle A
radius_B = 5   # Radius for particle B
time_step = 0.1
temperature = 300  # Desired temperature
reactivity_rate = 0.5  # Define the reactivity rate (0 to 1)

# Initialize screen
screen = pygame.display.set_mode((box_width, box_height))
pygame.display.set_caption("Hard Sphere Molecular Dynamics")

# Colors
white = (255, 255, 255)
blue = (0, 0, 255)
green = (0, 255, 0)
red = (255, 0, 0)  # Color for particle C

# Maxwell-Boltzmann distribution function for velocities
def maxwell_boltzmann(v, m, T):
    return (m / (2 * np.pi * 1.38e-23 * T)) ** 0.5 * np.exp(-m * v ** 2 / (2 * 1.38e-23 * T))

# Initialize particles with random positions and velocities based on Maxwell-Boltzmann distribution
particles = []
for i in range(num_particles):
    if i < num_particles // 2:  # First half of particles are 'A'
        radius = radius_A
        color = blue
    else:  # Second half of particles are 'B'
        radius = radius_B
        color = green
        
    x = np.random.uniform(radius, box_width - radius)
    y = np.random.uniform(radius, box_height - radius)
    # Generate velocities from Maxwell-Boltzmann distribution
    vx = np.random.normal(0, np.sqrt(temperature / 2))
    vy = np.random.normal(0, np.sqrt(temperature / 2))
    particles.append({'position': np.array([x, y]),
                      'velocity': np.array([vx, vy]),
                      'radius': radius,
                      'color': color,
                      'type': 'A' if i < num_particles // 2 else 'B'})

def update_particles():
    for particle in particles:
        particle['position'] += particle['velocity'] * time_step

def handle_wall_collisions():
    for particle in particles:
        if particle['position'][0] <= particle['radius'] or particle['position'][0] >= box_width - particle['radius']:
            particle['velocity'][0] *= -1
        if particle['position'][1] <= particle['radius'] or particle['position'][1] >= box_height - particle['radius']:
            particle['velocity'][1] *= -1

def handle_particle_collisions():
    for i in range(len(particles)):
        for j in range(i + 1, len(particles)):
            particle_i = particles[i]
            particle_j = particles[j]
            distance = np.linalg.norm(particle_i['position'] - particle_j['position'])
            if distance <= particle_i['radius'] + particle_j['radius']:
                overlap = particle_i['radius'] + particle_j['radius'] - distance
                normal = (particle_i['position'] - particle_j['position']) / distance
                relative_velocity = particle_i['velocity'] - particle_j['velocity']
                impulse = 2 * np.dot(relative_velocity, normal) / (1 / particle_i['radius'] + 1 / particle_j['radius'])
                
                # Update velocities with the same data type as 'radius'
                particle_i['velocity'] = (particle_i['velocity'] - impulse / particle_i['radius'] * normal).astype(float)
                particle_j['velocity'] = (particle_j['velocity'] + impulse / particle_j['radius'] * normal).astype(float)
                
                # Move particles apart to eliminate overlap
                displacement = 0.5 * overlap * normal
                particle_i['position'] += displacement
                particle_j['position'] -= displacement


def handle_chemical_reaction():
    particles_to_remove = []  # Create a list to store indices of particles to remove
    for i in range(len(particles)):
        for j in range(i + 1, len(particles)):
            particle_i = particles[i]
            particle_j = particles[j]
            distance = np.linalg.norm(particle_i['position'] - particle_j['position'])
            
            # Check if particles are of different types ('A' and 'B')
            if particle_i['type'] == 'A' and particle_j['type'] == 'B' and distance <= particle_i['radius'] + particle_j['radius']:
                # Determine if a reaction occurs based on reactivity_rate
                if random.random() < reactivity_rate:
                    # Create a new particle 'C' at the midpoint of 'A' and 'B'
                    new_x = (particle_i['position'][0] + particle_j['position'][0]) / 2
                    new_y = (particle_i['position'][1] + particle_j['position'][1]) / 2
                    new_radius = particle_i['radius'] + particle_j['radius']
                    # Assign an initial velocity to particle 'C' (modify as needed)
                    new_velocity = np.array([random.uniform(-1, 1), random.uniform(-1, 1)])
                    particles.append({'position': np.array([new_x, new_y]),
                                      'velocity': new_velocity,  # Initial velocity of 'C'
                                      'radius': new_radius,
                                      'color': red,
                                      'type': 'C'})
                        
                    # Mark 'A' and 'B' indices for removal
                    particles_to_remove.extend([i, j])

    # Remove particles by index in reverse order to avoid shifting indices
    particles_to_remove = list(set(particles_to_remove))  # Remove duplicates
    particles_to_remove.sort(reverse=True)
    for idx in particles_to_remove:
        particles.pop(idx)

def draw_particles():
    screen.fill(white)
    for particle in particles:
        pygame.draw.circle(screen, particle['color'], particle['position'].astype(int), particle['radius'])
        
# Define a function to calculate derivatives of a list
def calculate_derivative(data):
    return np.diff(data)

count_A = []
count_B = []
count_C = []

def main():
    running = True
    velocities = []  # List to store velocities for histogram

    frame = 0  # Initialize frame counter

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        update_particles()
        handle_wall_collisions()
        handle_particle_collisions()
        handle_chemical_reaction()  # Handle chemical reactions
        draw_particles()
        pygame.display.flip()

        # Count the number of A, B, and C particles at each frame
        num_A = sum(1 for particle in particles if particle['type'] == 'A')
        num_B = sum(1 for particle in particles if particle['type'] == 'B')
        num_C = sum(1 for particle in particles if particle['type'] == 'C')
        
        count_A.append(num_A)
        count_B.append(num_B)
        count_C.append(num_C)

        # Collect velocities for histogram
        velocities.extend([np.linalg.norm(p['velocity']) for p in particles])

        frame += 1  # Increment frame counter

    # Calculate the derivatives of A, B, and C counts
    derivative_A = calculate_derivative(count_A)
    derivative_B = calculate_derivative(count_B)
    derivative_C = calculate_derivative(count_C)
    
    if len(derivative_A) > 0 and len(derivative_B) > 0 and len(derivative_C) > 0:
        # Perform FFT calculations only if there are data points
        fft_A = np.fft.fft(derivative_A)
        fft_B = np.fft.fft(derivative_B)
        fft_C = np.fft.fft(derivative_C)

        # Calculate the sampling rate
        sampling_rate = 1  # Assuming one frame per time step

        # Calculate the frequencies
        freq_A = np.fft.fftfreq(len(derivative_A), 1 / sampling_rate)
        freq_B = np.fft.fftfreq(len(derivative_B), 1 / sampling_rate)
        freq_C = np.fft.fftfreq(len(derivative_C), 1 / sampling_rate)
        
        fft_A = np.fft.fft(derivative_A)
        fft_B = np.fft.fft(derivative_B)
        fft_C = np.fft.fft(derivative_C)
        
        # Calculate the sampling rate
        sampling_rate = 1  # Assuming one frame per time step

        # Calculate the frequencies
        freq_A = np.fft.fftfreq(len(derivative_A), 1 / sampling_rate)
        freq_B = np.fft.fftfreq(len(derivative_B), 1 / sampling_rate)
        freq_C = np.fft.fftfreq(len(derivative_C), 1 / sampling_rate)
        
        
    plt.figure(figsize=(10, 6))
    plt.plot(freq_A, np.abs(fft_A), label='FFT of dA/dt', color='blue', linestyle='-', linewidth=2)
    plt.plot(freq_B, np.abs(fft_B), label='FFT of dB/dt', color='green', linestyle='-', linewidth=2)
    plt.plot(freq_C, np.abs(fft_C), label='FFT of dC/dt', color='red', linestyle='-', linewidth=2)
    plt.xlabel('Frequency (cycles per frame)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()


    # Create and display a graph of the derivatives of A, B, and C counts
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, frame), derivative_A, label="dA/dt", color="blue", linestyle="-", linewidth=5)
    plt.plot(range(1, frame), derivative_B, label="dB/dt", color="green", linestyle="-", linewidth=5)
    plt.plot(range(1, frame), derivative_C, label="dC/dt", color="red", linestyle="-", linewidth=5)
    plt.xlabel("Frame")
    plt.ylabel("Derivative")
    plt.legend()
    plt.savefig("derivative_ABC.png", dpi=600)
    plt.show()
    # Create and display histogram plot
    plt.hist(velocities, bins=20, density=True, alpha=0.7, color='blue')
    plt.xlabel('Velocity')
    plt.ylabel('Frequency')
    plt.title('Particles Velocity Distribution')

    # Fit a Maxwell-Boltzmann distribution to the histogram
    params = stats.maxwell.fit(velocities)
    x = np.linspace(0, np.max(velocities), 100)
    fitted_pdf = stats.maxwell.pdf(x, *params)
    plt.plot(x, fitted_pdf, color='red', label='Maxwell-Boltzmann Fit')
    plt.savefig('MBplot.png')
    plt.legend()
    plt.show()
    
    # Create and display a graph of A, B, and C particles over time
    plt.figure(figsize=(10, 6))  # Optional: Set the figure size
    plt.plot(range(frame), count_A, label='A', color='blue', linestyle='-', linewidth=5)
    plt.plot(range(frame), count_B, label='B', color='green', linestyle='--', linewidth=5)  # Change linestyle to '--'
    plt.plot(range(frame), count_C, label='C', color='red', linestyle='-', linewidth=5)
    plt.xlabel('Frame')
    plt.ylabel('Count')
    plt.savefig('reaction.png', dpi = 600)
    plt.legend()
    plt.show()
    
    
    # Create and display a half-life graph for particle A
    plt.figure(figsize=(10, 6))
    plt.plot(range(frame), count_A, label='A', color='blue', linestyle='-', linewidth=5)
    plt.axhline(max(count_A) / 2, color='blue', linestyle='--', label='Half-Life of A')
    plt.xlabel('Frame')
    plt.ylabel('Count')
    plt.savefig('meia_vida_a.png', dpi = 600)
    plt.legend()
    plt.show()

    # Create and display a half-life graph for particle B
    plt.figure(figsize=(10, 6))
    plt.plot(range(frame), count_B, label='B', color='green', linestyle='-', linewidth=5)
    plt.axhline(max(count_B) / 2, color='green', linestyle='--', label='Half-Life of B')
    plt.xlabel('Frame')
    plt.ylabel('Count')
    plt.savefig('meia_vida_b.png', dpi = 600)
    plt.legend()
    plt.show()

    # Create and display a half-life graph for particle C
    plt.figure(figsize=(10, 6))
    plt.plot(range(frame), count_C, label='C', color='red', linestyle='-', linewidth=5)
    plt.axhline(max(count_C) / 2, color='red', linestyle='--', label='Half-Life of C')
    plt.xlabel('Frame')
    plt.ylabel('Count')
    plt.savefig('meia_vida_c.png', dpi = 600)
    plt.legend()
    plt.show()
    
if __name__ == "__main__":
    main()
```
